```python
from sympy import init_printing
init_printing() 
from sympy import init_session
init_session() 
from sympy import *
from IPython.display import display
```

```python
import sympy as sp

#This is the correct solution for Fsb:

j, k, l, RT, LT, LsT, Fsb, Kd1, Kd2, Kd3 = sp.symbols('j k l RT LT LsT Fsb Kd1 Kd2 Kd3')

# Define j, k, l
j = LsT*Fsb**2 - (Kd1+LsT+RT)*Fsb + RT
k = LsT*Fsb**2 - (Kd3+LsT+RT)*Fsb + RT
l = LsT*Fsb - Kd3 - LsT

# Define the main equation
eq = sp.Eq(LT , (j*((k*l-(1-Fsb)*Kd2*Kd3)*Kd1+(1-Fsb)*Kd2*Kd3**2)) / (k*(1-Fsb)*(Kd1-Kd3)*Kd1))
display(eq)

# Multiply both sides by the denominator to clear fractions
denominator = (k*(1-Fsb)*(Kd1-Kd3)*Kd1)
eq2 = sp.Eq(eq.lhs*denominator, eq.rhs*denominator)
display(eq2)

eq3 = sp.expand(eq2)
display(eq3)

polynom = eq3.rhs - eq3.lhs
print("polynom:")
display(polynom)

polynom_collected = sp.collect(polynom, Fsb)
print("polynom collected for Fsb:")
display(polynom_collected)

# Get the coefficients
coeffs = [polynom_collected.coeff(Fsb, i) for i in range(6)]

# Print the coefficients
for i, coeff in enumerate(coeffs):
    simplified_coeff = sp.simplify(coeff)
    print(f"Coefficient of Fsb^{i}: {simplified_coeff}")

print()

# Display the coefficients
for i, coeff in enumerate(coeffs):
    simplified_coeff = sp.simplify(coeff)
    print(f"Coefficient of Fsb^{i}:")
    display(simplified_coeff)
```

```python

# This is the solution for R:

j, k, l, R, RT, LT, LsT, Fsb, Kd1, Kd2, Kd3 = sp.symbols('j k l R RT LT LsT Fsb Kd1 Kd2 Kd3')

Fsb = R / (Kd1 + R)

# Define j, k, l
j = LsT*Fsb**2 - (Kd1+LsT+RT)*Fsb + RT
k = LsT*Fsb**2 - (Kd3+LsT+RT)*Fsb + RT
l = LsT*Fsb - Kd3 - LsT

# Define the main equation
eq = sp.Eq(LT , (j*((k*l-(1-Fsb)*Kd2*Kd3)*Kd1+(1-Fsb)*Kd2*Kd3**2)) / (k*(1-Fsb)*(Kd1-Kd3)*Kd1))
display(eq)

# Multiply both sides by the denominator to clear fractions
denominator = (k*(1-Fsb)*(Kd1-Kd3)*Kd1)
eq2 = sp.Eq(eq.lhs*denominator, eq.rhs*denominator)
display(eq2)

polynom = eq2.rhs - eq2.lhs
print("polynom:")
print(polynom)
display(polynom)

common_denominator = (Kd1+R)**2
polynom = sp.simplify(polynom * common_denominator) * (Kd1+R)**3
print("polynom:")
display(polynom)

polynom = sp.cancel(polynom)
print("polynom cancelled:")
display(polynom)


polynom_collected = sp.collect(polynom, R)
print("polynom collected for R:")
display(polynom_collected)


# Get the coefficients
coeffs = [polynom_collected.coeff(R, i) for i in range(6)]

# Print the coefficients
for i, coeff in enumerate(coeffs):
    simplified_coeff = sp.simplify(coeff)
    print(f"Coefficient of R^{i}: {simplified_coeff}")

print()

# Display the coefficients
for i, coeff in enumerate(coeffs):
    simplified_coeff = sp.simplify(coeff)
    print(f"Coefficient of R^{i}:")
    display(simplified_coeff)
```
