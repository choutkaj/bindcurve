# Component Playground

This notebook is a pure playground for the `dir_specific` and `comp_3st_specific` models. Nothing is fitted and no data are loaded. You just edit model parameters, rerun the plotting cells, and inspect how the hidden system components move.

```python
import numpy as np
import matplotlib.pyplot as plt

import bindcurve as bc

plot_floor = 1.0e-6
direct_specific_model = bc.DirectSpecificKdModel()
comp_3st_specific_model = bc.CompetitiveThreeStateSpecificKdModel()
```

## How to use this notebook

1. Edit one of the parameter dictionaries below.
2. Rerun the matching plotting cell.
3. Watch how the component curves and the observable model response change.

The component panels use log scales on both axes so species living on very different concentration scales stay visible. A tiny positive floor is only used for plotting, because log axes cannot display exact zeros.

```python
def log_grid(x_min: float, x_max: float, n_points: int = 400) -> np.ndarray:
    return np.logspace(np.log10(x_min), np.log10(x_max), n_points)


def positive_for_plot(values: np.ndarray, floor: float = plot_floor) -> np.ndarray:
    return np.clip(np.asarray(values, dtype=float), floor, None)


def direct_specific_components(
    receptor_total: np.ndarray,
    *,
    ymin: float,
    ymax: float,
    LsT: float,
    Kds: float,
) -> dict[str, np.ndarray]:
    receptor_total = np.asarray(receptor_total, dtype=float)
    a = Kds + LsT - receptor_total
    b = -Kds * receptor_total
    receptor_free = (-a + np.sqrt(np.maximum(a**2 - 4.0 * b, 0.0))) / 2.0
    tracer_bound = np.clip(receptor_total - receptor_free, 0.0, None)
    tracer_free = np.clip(LsT - tracer_bound, 0.0, None)
    response = direct_specific_model.evaluate(
        receptor_total,
        ymin=ymin,
        ymax=ymax,
        LsT=LsT,
        Kds=Kds,
    )
    return {
        'x': receptor_total,
        'R_total': receptor_total,
        'R_free': receptor_free,
        'Lstar_total': np.full_like(receptor_total, LsT, dtype=float),
        'Lstar_free': tracer_free,
        'RLstar': tracer_bound,
        'response': np.asarray(response, dtype=float),
    }


def competitive_three_state_components(
    ligand_total: np.ndarray,
    *,
    ymin: float,
    ymax: float,
    RT: float,
    LsT: float,
    Kds: float,
    Kd: float,
) -> dict[str, np.ndarray]:
    ligand_total = np.asarray(ligand_total, dtype=float)
    a = Kds + Kd + LsT + ligand_total - RT
    b = Kds * (ligand_total - RT) + Kd * (LsT - RT) + Kds * Kd
    c = -Kds * Kd * RT

    discriminant = np.maximum(a**2 - 3.0 * b, 0.0)
    denominator = 2.0 * np.sqrt(discriminant**3)
    numerator = -2.0 * a**3 + 9.0 * a * b - 27.0 * c
    argument = np.divide(
        numerator,
        denominator,
        out=np.zeros_like(numerator, dtype=float),
        where=denominator > 0.0,
    )
    theta = np.arccos(np.clip(argument, -1.0, 1.0))
    receptor_free = -(a / 3.0) + (2.0 / 3.0) * np.sqrt(discriminant) * np.cos(theta / 3.0)

    tracer_bound = np.clip(LsT * receptor_free / (Kds + receptor_free), 0.0, None)
    tracer_free = np.clip(LsT - tracer_bound, 0.0, None)
    competitor_bound = np.clip(ligand_total * receptor_free / (Kd + receptor_free), 0.0, None)
    competitor_free = np.clip(ligand_total - competitor_bound, 0.0, None)
    response = comp_3st_specific_model.evaluate(
        ligand_total,
        ymin=ymin,
        ymax=ymax,
        RT=RT,
        LsT=LsT,
        Kds=Kds,
        Kd=Kd,
    )
    return {
        'x': ligand_total,
        'L_total': ligand_total,
        'R_total': np.full_like(ligand_total, RT, dtype=float),
        'R_free': receptor_free,
        'Lstar_total': np.full_like(ligand_total, LsT, dtype=float),
        'Lstar_free': tracer_free,
        'L_free': competitor_free,
        'RLstar': tracer_bound,
        'RL': competitor_bound,
        'response': np.asarray(response, dtype=float),
    }


def plot_dir_specific_playground(params: dict[str, float]) -> None:
    x = log_grid(params['x_min'], params['x_max'], int(params.get('n_points', 400)))
    components = direct_specific_components(x, **{
        'ymin': params['ymin'],
        'ymax': params['ymax'],
        'LsT': params['LsT'],
        'Kds': params['Kds'],
    })

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

    axes[0].plot(x, positive_for_plot(components['R_total']), linestyle='--', color='0.55', linewidth=1.5, label='[R]_T')
    axes[0].plot(x, positive_for_plot(components['R_free']), color='#1f77b4', linewidth=2, label='[R]')
    axes[0].plot(x, positive_for_plot(components['Lstar_total']), linestyle=':', color='0.35', linewidth=1.5, label='[L*]_T')
    axes[0].plot(x, positive_for_plot(components['Lstar_free']), color='#2ca02c', linewidth=2, label='[L*]')
    axes[0].plot(x, positive_for_plot(components['RLstar']), color='#d62728', linewidth=2, label='[RL*]')
    axes[0].set_xscale('log')
    axes[0].set_yscale('log')
    axes[0].set_xlabel('receptor total')
    axes[0].set_ylabel('component concentration')
    axes[0].set_title('dir_specific components')
    axes[0].legend(ncol=2)

    axes[1].plot(x, components['response'], color='tab:blue', linewidth=2)
    axes[1].set_xscale('log')
    axes[1].set_xlabel('receptor total')
    axes[1].set_ylabel('observable response')
    axes[1].set_title('dir_specific observable curve')

    plt.tight_layout()
    plt.show()


def plot_comp_3st_specific_playground(params: dict[str, float]) -> None:
    x = log_grid(params['x_min'], params['x_max'], int(params.get('n_points', 400)))
    components = competitive_three_state_components(x, **{
        'ymin': params['ymin'],
        'ymax': params['ymax'],
        'RT': params['RT'],
        'LsT': params['LsT'],
        'Kds': params['Kds'],
        'Kd': params['Kd'],
    })

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

    axes[0].plot(x, positive_for_plot(components['L_total']), linestyle='--', color='0.55', linewidth=1.5, label='[L]_T')
    axes[0].plot(x, positive_for_plot(components['R_total']), linestyle=':', color='0.35', linewidth=1.5, label='[R]_T')
    axes[0].plot(x, positive_for_plot(components['Lstar_total']), linestyle='-.', color='0.45', linewidth=1.5, label='[L*]_T')
    axes[0].plot(x, positive_for_plot(components['R_free']), color='#1f77b4', linewidth=2, label='[R]')
    axes[0].plot(x, positive_for_plot(components['Lstar_free']), color='#2ca02c', linewidth=2, label='[L*]')
    axes[0].plot(x, positive_for_plot(components['L_free']), color='#9467bd', linewidth=2, label='[L]')
    axes[0].plot(x, positive_for_plot(components['RLstar']), color='#d62728', linewidth=2, label='[RL*]')
    axes[0].plot(x, positive_for_plot(components['RL']), color='#8c564b', linewidth=2, label='[RL]')
    axes[0].set_xscale('log')
    axes[0].set_yscale('log')
    axes[0].set_xlabel('competitor total')
    axes[0].set_ylabel('component concentration')
    axes[0].set_title('comp_3st_specific components')
    axes[0].legend(ncol=2)

    axes[1].plot(x, components['response'], color='tab:orange', linewidth=2)
    axes[1].set_xscale('log')
    axes[1].set_xlabel('competitor total')
    axes[1].set_ylabel('observable response')
    axes[1].set_title('comp_3st_specific observable curve')

    plt.tight_layout()
    plt.show()
```

## `dir_specific` playground

Edit the dictionary below and rerun the next cell. The default `ymin=0` and `ymax=1` make the observable response numerically equal to tracer-bound fraction.

```python
dir_specific_params = {
    'x_min': 1.0e-4,
    'x_max': 1.0e1,
    'n_points': 400,
    'ymin': 0.0,
    'ymax': 1.0,
    'LsT': 0.35,
    'Kds': 1.8,
}

dir_specific_params
```

```python
plot_dir_specific_playground(dir_specific_params)
```

## `comp_3st_specific` playground

Edit the dictionary below and rerun the next cell. Try changing `Kd`, `Kds`, `RT`, or `LsT` one at a time to see which species are most sensitive.

```python
comp_3st_specific_params = {
    'x_min': 1.0e-4,
    'x_max': 1.0e1,
    'n_points': 400,
    'ymin': 0.0,
    'ymax': 1.0,
    'RT': 0.05,
    'LsT': 0.005,
    'Kds': 0.02,
    'Kd': 1.6,
}

comp_3st_specific_params
```

```python
plot_comp_3st_specific_playground(comp_3st_specific_params)
```

If you want to compare multiple parameter sets, duplicate one of the dictionaries, rename it, and call the plotting helper again.
