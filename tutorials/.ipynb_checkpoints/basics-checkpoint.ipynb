# Basics of BindCurve

In this tutorial, we will go through the very basics of `bindcurve`. We will fit IC50 for six compounds and plot the results. This will demonstrate the most minimalistic workflow you can do to get your results quickly and easily. For clarity, we will mostly use default settings. There is much more you can do, please see more advanced tutorials for more detailed overview of the available options and settings.

## Preparations

First, let's start by importing `bindcurve`. We will use `bc` as alias to help with clarity. Also, we will import pandas and matplotlib because we will need it. Don't worry, all dependencies have been installed automatically when installing `bindcurve`.

```python
import bindcurve as bc
import pandas as pd
import matplotlib.pyplot as plt
```

The following settings adjust the decimals and display width for pandas. This is just for display purposes, all numbers will still be stored with many decimal places.

```python
pd.set_option("display.precision", 2)
pd.set_option('display.width', 200)
```

## Load the data

With all preparations taken care of, let's load our data from csv file. You can find this data on GitHub in the folder `/docs/tutorials/examples/competitive.csv`. As you can see, the loaded data have been preprocessed and now newly contain log-transformed concentrations (log c) as well as number of replicates (n_reps) and median values with SD snd SEM metrics for each concentration point.

```python
input_data = bc.load_csv("examples/competitive.csv")

print(input_data)
```

We can also get the list of compounds in our data and print the number of them.

```python
compounds = input_data["compound"].unique()
print("Detected compounds:", compounds)
print("No. of detected compounds:", len(compounds))
```

## Fitting the model

Finally, we can fit a logistic model into our data. We will do that by using the `fit_50` function and we will pass the DataFrame with the input data, and the model name, as the arguments for this function. The function will output a new Dataframe containing the results. The results contain the following information:

- `n_points`: number of datapoints that were used for fitting
- `IC50`: value of IC50
- `loCL`: lower confidence limit of the 95% confidence interval
- `upCL`: upper confidence limit of the 95% confidence interval
- `SE`: standard error
- `model`: name of the model that was used
- `ymin`: lower asymptote of the model
- `ymax`: upper asymptote of the model
- `slope`: slope of the model
- `Chi^2`: estimate of $\chi^2$ (the lower the better)
- `R^2`: estimate of $R^2$ (the closer to 1 the better)

```python
IC50_results = bc.fit_50(input_data, model="IC50")
print(IC50_results)
```

We can also get the results as a formatted report using the `report` function. This will provide a convenient report in two formats:
- mean (95% confidence interval)
- mean ± SE

The last line will copy the report into your clipboard, you can then paste the results by CTRL+V anywhere you want.

```python
IC50_report = bc.report(IC50_results, decimals=1)
print(IC50_report)
IC50_report.to_clipboard(excel=True, sep=",")
```

## Plotting the curves

Plotting is done using the matplotlib library. We will initiate the plot in a standard way, then we will use the `plot` function to actually pass the curves into the plot. Then we will use common matplotlib settings to set up title, axis labels, and set x axis to log scale. This way you have a lot of flexibility in designing your plot.  

```python
plt.figure(figsize=(5, 4))

bc.plot(input_data, IC50_results)

plt.title("Title of my plot")
plt.xlabel("Concentration")
plt.ylabel("Response")
plt.xscale("log")
plt.legend()

# Uncomment the below line if you want to save your plot as png
#plt.savefig("my_plot.png", dpi=300) 

plt.show()
```

We can also plot just a selected compounds using the `compound_sel` argument. The selection is supplied as a list of compound names. We will also show all the datapoints instead of onyl medians using `show_all_data=True` and `show_medians=False` arguments. We will also hide the errorbars by using `show_errorbars=False`. Let's also make the markers smaller to get better clarity, we will accomplish that by `markersize=3`.

```python
plt.figure(figsize=(5, 4))

bc.plot(input_data, 
        IC50_results, 
        compound_sel=["comp3", "comp6"], 
        show_all_data=True, 
        show_medians=False, 
        show_errorbars=False, 
        markersize=3)

plt.title("Plotting selection")
plt.xlabel("Concentration")
plt.ylabel("Response")
plt.xscale("log")
plt.legend()
plt.show()
```

Finally, you can also plot the compounds on a grid using the `plot_grid` function. To do this, you don't initiate the matplotlib plot prior to using the bindcurve function as in tha case of the `plot` function. Instead, everything is executed by a single line of code. You have to supply all the settings as arguments into the `plot_grid` function.

```python
bc.plot_grid(input_data, IC50_results, n_cols=3, figsize=(10,6))
```
