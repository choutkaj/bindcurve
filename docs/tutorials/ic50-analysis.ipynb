# IC₅₀ analysis

This tutorial fits two synthetic inhibitory dose-response datasets. Each compound has two independent experiments and three technical replicates per concentration. Concentrations are unit-agnostic but use one consistent scale. See the [logistic-model theory](../theory/logistic.md) for the model definition.

```python
import matplotlib.pyplot as plt

import bindcurve as bc
```

## Load the observations

Wide input stores technical replicates in separate response columns. `DoseResponseData` converts them to the canonical long-form representation.

```python
all_data = bc.DoseResponseData.from_csv(
    "data/competitive-binding.csv",
    format="wide",
)
data = all_data.keep_only(["ic50_a", "ic50_b"])
data.summary()
```

## Fit and summarize

`bindcurve` averages technical replicates at each concentration, fits each independent experiment separately, and then summarizes the fitted parameters across experiments. Here the known lower response and amplitude are fixed so the fit focuses on IC₅₀ and Hill slope.

```python
results = bc.fit(
    data,
    model="ic50",
    fixed={"ymin": 5.0, "amplitude": 90.0},
)

results.summary()[
    ["compound_id", "N_exp", "N_fit_successful", "IC50", "hill_slope"]
]
```

```python
results.report(unit="concentration units", include_n_exp=True)
```

## Plot compound summaries

`plot_compounds()` shows one aggregate series per compound while preserving the experiment-level fits used for the summary.

```python
fig, ax = plt.subplots(figsize=(6, 4))
bc.plot_compounds(data, results, ax=ax)
ax.set_xlabel("concentration")
ax.set_ylabel("response")
ax.legend()
plt.show()
```
